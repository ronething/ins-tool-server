```python
#coding=utf-8
import requests
import re
import json
```

```python
proxies = {
    'http': 'socks5://host:port',
    'https': 'socks5://host:port',
}
headers = {
    'Referer': 'https://www.instagram.com/',
    'User-Agent': '',
}
```

```python
test_url = 'https://www.instagram.com/p/BwLD6O9BV07/?utm_source=ig_web_copy_link'
```

```python
def r1(pattern, text):
    m = re.search(pattern, text)
    if m:
        return m.group(1)
```

```python
url = r1(r'([^?]*)', test_url)
```

```python
url
```

```python
def get_html(url):
    try:
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            return response.text
        else:
            print('请求网页源代码错误, 错误状态码：', response.status_code)
    except Exception as e:
        print(e)
        return None
```

```python
html = get_html(url)
```

```python
html
```

```python
vid = r1(r'instagram.com/p/([^/]+)', url)
description = r1(r'<meta property="og:title" content="([^"]*)"', html)
title = "{} [{}]".format(description.replace("\n", " "), vid)
stream = r1(r'<meta property="og:video" content="([^"]*)"', html)
```
